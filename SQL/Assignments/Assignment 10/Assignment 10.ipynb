# Dice

#### Query 1: Before

```python
# dice before
_, df = dataframe_query(conn, 
    """
    SELECT production_location.*, rating_fact.*, movie_fact.*
    FROM rating_fact, movie_fact, production_location, movie
    WHERE rating_fact.Movie_ID = movie.Movie_ID
    AND movie.Movie_ID = movie_fact.Movie_ID
    AND production_location.Country_Code = movie_fact.Country_Code
    """
)

df
```

#### Query 1: After

```python
# dice after
# production_location in USA or France and rating between 4 and 9
_, df = dataframe_query(conn, 
    """
    SELECT production_location.*, rating_fact.*, movie_fact.*
    FROM rating_fact, movie_fact, production_location
    WHERE production_location.Country_Code IN ('USA', 'FRA')
    AND rating_fact.rating BETWEEN 4 AND 9
    AND rating_fact.Movie_ID = movie_fact.Movie_ID
    AND production_location.Country_Code = movie_fact.Country_Code
    """
)

df
```

#### Query 2: Before

```python
# dice before
_, df = dataframe_query(conn, 
    """
    SELECT production_location.*, rating_fact.*, movie_fact.*
    FROM production_location, rating_fact, movie_fact
    WHERE rating_fact.Movie_ID = movie_fact.Movie_ID
    AND production_location.Country_Code = movie_fact.Country_Code
    """
)

df
```

#### Query 2: After

```python
# dice after
# production_location in continents of Europe or Asia and movie ID as 710 or 8844
_, df = dataframe_query(conn, 
    """
    SELECT production_location.*, rating_fact.*, movie_fact.*
    FROM production_location, rating_fact, movie_fact
    WHERE production_location.Continent IN ("Europe", 'Asia')
    AND movie_fact.Movie_ID IN ('710', '8844')
    AND rating_fact.Movie_ID = movie_fact.Movie_ID
    AND production_location.Country_Code = movie_fact.Country_Code
    """
)

df
```

# Drill Down

#### Query 1: Before 

```python
_, df = dataframe_query(conn, 
    """
    SELECT date.Year, date.Month, AVG(movie_fact.Revenue) as Average
    FROM date, production_location, movie_fact
    WHERE movie_fact.Country_Code = production_location.Country_Code
    AND   movie_fact.FullDate = date.FullDate
    GROUP BY date.Year, date.Month
    """
)

df
```

#### Query 1: After

```python
# drill down to Day
_, df = dataframe_query(conn, 
    """
    SELECT date.Year, date.Month, date.Day, AVG(movie_fact.Revenue) as Average
    FROM date, production_location, movie_fact
    WHERE movie_fact.Country_Code = production_location.Country_Code
    AND   movie_fact.FullDate = date.FullDate
    GROUP BY date.Year, date.Month, date.Day
    """
)

df
```

#### Query 2: Before

```python
_, df = dataframe_query(conn, 
    """
    SELECT p.Continent, p.Country, AVG(movie_fact.Revenue) AS Average_Revenue
    FROM production_location p, movie_fact   
    WHERE movie_fact.Country_Code = p.Country_Code
    GROUP BY Continent, Country
    ORDER BY Continent, Country
    """
)

df
```

#### Query 2: After

```python
# Drill down to City
_, df = dataframe_query(conn, 
    """
    SELECT p.Continent, p.Country, p.City, AVG(movie_fact.Revenue) AS Average_Revenue
    FROM production_location p, movie_fact
    WHERE movie_fact.Country_Code = p.Country_Code
    GROUP BY Continent, Country, City
    ORDER BY Continent, Country, City
    """
)

df
```

# Roll Up

```python
# Using at least one of your dimensions, perform a query that uses a GROUP BY ... WITH ROLLUP 
# operation. Display the result in a dataframe. You can simply display the dataframe with the None cells.
```

```python
_, df = dataframe_query(conn, 
    """
    SELECT date.Year, date.Month, date.Day, AVG(movie_fact.Revenue) as Average
    FROM date, production_location, movie_fact
    WHERE movie_fact.Country_Code = production_location.Country_Code
    AND   movie_fact.FullDate = date.FullDate
    GROUP BY date.Year, date.Month, date.Day
    WITH ROLLUP
    """
)

df
```
